```python
import numpy as np 
import pandas as pd 
import sqlite3 as sql

```

```python
db = 'movies1.sqlite'
con = sql.connect(db) 
cur = con.cursor()
```

```python
cur.execute('SELECT * FROM directors')
directors = cur.fetchall()
directors = pd.DataFrame(directors, columns = ['name', 'id', 'gender', 'uid', 'department'])
directors.head()
```

```python
cur.execute('SELECT * FROM movies')
movies = cur.fetchall()
movies = pd.DataFrame(movies, columns = ['id', 'original_title', 'budget', 'popularity', 'release_date', 'revenue', 'title', 'vote_average', 'vote_count', 'overview', 'tagline', 'uid', 'director_id'])
movies.head(3)
```

```python
# Q3
query1 = """
    SELECT COUNT(*) AS count
    FROM directors
    Where gender=='1';
"""
directors_female = pd.read_sql(query1, con)
query2 = """
    SELECT COUNT(*) AS count
    FROM directors
    Where gender=='2';
"""
directors_male = pd.read_sql(query2, con)
ans=directors_male/directors_female

print("Ratio of male to female directors is ",ans['count'].sum(),":1")
```
